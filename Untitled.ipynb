```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load and Inspect
df = pd.read_csv("fintech_data.csv")
print(df.head(10))
print(df.dtypes)
```

```python
#Data Cleaning
df["Date"] = pd.to_datetime(df["Date"])
df["Transaction_Amount"] = df["Transaction_Amount"].astype(int)
df["Transaction_Fee"] = df["Transaction_Fee"].astype(int)
df["Net_Amount"] = df["Net_Amount"].astype(int)
df.head(3)
```

```python
#Descriptive Analysis
print("What is the average transaction fee?")
print(df["Transaction_Fee"].mean())
print("---")
print("What’s the total revenue from fees collected?")
print(df["Transaction_Fee"].sum())
print("---")
print("Which city processed the most transactions?")
print(df["City"].value_counts().head(2).sort_values(ascending=False))
print("---")
print("Which service type had the highest total transaction volume?")
print(df["Service_Type"].value_counts().head(1).sort_values(ascending=False))
```

```python
#Grouping and Aggregation
print("Total and average transaction amount by City.")
ta = df.groupby(["City"])["Transaction_Amount"].mean()
print(ta)
print("---")
print("Total net amount grouped by Branch_ID.")
tn = df.groupby(["Branch_ID"])["Net_Amount"].sum()
print(tn)
print("---")
print("Count of transactions for each Service_Type.")
print(df["Service_Type"].value_counts())
```

```python
#Visualizations
print("Bar chart: Total Transaction_Amount per city.")
tm = df.groupby(["City"])["Transaction_Amount"].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
tm.plot.bar(title="Total Transaction_Amount per city")
plt.ylabel("Transaction_Amount")
plt.xlabel("City")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
print("---")
print("Pie chart: Distribution of Account_Type.")
at = df["Account_Type"].value_counts()
at.plot(kind="pie", autopct='%1.1f%%', ylabel='Distribution of Account_Type')
plt.show()
print("---")
print("Line chart: Daily transaction trend over time.")
dt = df.groupby(["Date"])["Transaction_Amount"].sum()
plt.figure(figsize=(12,6))
dt.plot(kind="line", color="blue")
plt.ylabel("Total Transaction Amount")
plt.xlabel("Date")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Bonus Analysis
print("What is the most commonly used transaction Channel?")
print(df["Channel"].value_counts().sort_values(ascending=False))
print("---")
print("Which currency (NGN or USD) was used more frequently?")
print(df["Currency"].value_counts().sort_values(ascending=False))
print("---")
print("Which Branch_ID had the highest average net amount per transaction?")
print(df.groupby(["Branch_ID"])["Net_Amount"].mean().sort_values(ascending=False))
```
